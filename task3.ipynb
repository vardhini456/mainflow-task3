```python
import pandas as pd

# Sample DataFrame
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [150, 200, 300, 250, 400, 450],
    'Profit': [50, 60, 80, 70, 100, 120]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Data
months = df['Month']
sales = df['Sales']
profit = df['Profit']

# Plotting the Bar Chart
plt.figure(figsize=(10, 5))

plt.bar(months, sales, color='skyblue', label='Sales')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()

plt.show()

# Plotting the Line Chart
plt.figure(figsize=(10, 5))

plt.plot(months, profit, color='green', marker='o', linestyle='-', label='Profit')
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.legend()

plt.show()
```
